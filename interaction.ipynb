```python
import altair as alt
import matplotlib.pyplot as plt
import seaborn
import pandas as pd
```

```python
prediction = pd.read_csv('car_price_prediction_.csv')
```

```python
top5 = prediction.groupby('Brand').count().sort_values('Mileage',ascending=False).head(5)

top5_options = list(top5.index)
top5_options.insert(0,None)
top5_labels = list(top5.index)
top5_labels.insert(0,'All')
input_dropdown = alt.binding_select(options = top5_options,
                                   labels = top5_labels, 
                                   name = 'Make')

selection = alt.selection_point(fields = ['Brand'], bind = input_dropdown)

alt.Chart(prediction).mark_point().encode(
    x = 'Mileage:Q',
    y = 'Price:Q', 
    color = alt.condition(selection, alt.Color('Brand:N'), alt.value('lightgrey'))

).add_params(
    selection
)
```
